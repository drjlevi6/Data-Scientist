# whiten_isolated_pixels

### Whiten non-white pixels surrounded above, below, to left and right by white pixels.

input is a csv-grayscale file, reshaped if necessary to a 100x100 array of pixels, each represented by a single int of value 0-255.

```python
import numpy as np
import pandas as pd
from pandas import DataFrame
from matplotlib import pyplot as plt
from copy import deepcopy
```

```python
def whiten_csv(csv):
    '''Whiten "isolated" non-white pixels'''
    '''i.e., pixels whose neighbors above, below, left, right
        (where applicable) are white,'''
    ymax, xmax = csv.shape
    cxv_new = deepcopy(csv)
    
    # Whiten interior
    for i in range(1, ymax-1):
        for j in range(1, xmax-1):
            if cxv_new[i-1][j] == 255 and cxv_new[i+1][j] == 255 and \
                    cxv_new[i][j-1] == 255 and cxv_new[i][j+1] == 255:
                cxv_new[i][j] = 255
    return cxv_new
```

```python
path = "../textlines-20240217-1-whitened.csv"
```

```python
csv_ori = np.genfromtxt(path, delimiter=',', dtype=int)
csv_whitened = np.array(whiten_csv(csv_ori), dtype=int)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(csv_ori, cmap="gray")
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(csv_whitened, cmap="gray")
```
