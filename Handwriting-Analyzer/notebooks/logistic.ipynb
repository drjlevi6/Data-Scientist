```python
%matplotlib inline

import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt

from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_excel("../xlsx/xpm_dataset_copy.xlsx") 
df.head()
```

```python
# Extract pixel data
# df.shape # (64, 10002)
df_data_column_list = list(range(1,df.shape[1]-1))
df_pixel_data = df.iloc[:, df_data_column_list]
df_pixel_data.head()
```

```python
X = df_pixel_data.values
y = df.loc[:, 'Letter'].values
X, y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=3)
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
scaler = StandardScaler()

# Fit on training set only.
scaler.fit(X_train)

# Apply transform to both the training set and the test set.
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
clf = LogisticRegression()
```

```python
clf.fit(X_train, y_train)
```

### Letters: Predicted vs. Actual

```python
df_predictions = DataFrame(
    [(clf.predict(X_test[i].reshape(1,-1))[0],
    y_test[i]) for i in range(len(y_test))],
    columns = ["Predicted", "Actual"])
df_predictions
```
